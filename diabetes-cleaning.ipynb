```python
from scipy.stats import zscore
```

```python
df['Insulin_zscore']=zscore(df['Insulin'])
outliers=df[(df['Insulin_zscore']>3) | (df['Insulin_zscore']<-3)]
```

```python
outliers
```

```python
df['BMI']=zscore(df['BMI'])
outliers=df[(df['BMI']>3) | (df['BMI']<-3)]
```

```python
outliers
```

```python
Q1 = df['Insulin'].quantile(0.25)
Q3 = df['Insulin'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['Insulin'] < (Q1 - 1.5 * IQR)) | (df['Insulin']> (Q3 + 1.5 * IQR))]
```

```python
outliers
```

```python
#Handling / Removing Outliers
df_cleaned = df[(df['Insulin'] >= (Q1 - 1.5 * IQR)) & (df['Insulin'] <= (Q3 + 1.5 * IQR))]
df_cleaned
```
